```python
import pandas as pd
df = pd.read_csv('dataset1.csv')
display(df.head())
print(f"Size {df.shape}")
```

```python
df = pd.read_csv('dataset1.csv', header=None)
display(df.head())
print(f"Size {df.shape}")
```

```python
column_sum = df.sum(axis=0)
column_average = df.mean(axis=0)
row_sum = df.sum(axis=1)
row_average = df.mean(axis=1)
print(row_sum)
print(row_average)
print(column_sum)
print(column_average)
```

```python
df['Row sum'] = row_sum
df['Row Average'] = row_average
display(df)
```

```python
column_sum_row = pd.DataFrame(column_sum).T
column_average_row = pd.DataFrame(column_average).T
display(column_sum_row)
display(column_average_row)
```

```python
column_sum_row.index = ['Column_Sum']
column_average_row.index = ['Column_Average']
display(column_sum_row)
display(column_average_row)
```

```python
df_final = pd.concat([df, column_sum_row, column_average_row])
display(df_final)
```

```python
df_final.to_csv('update_dataset1.csv', index=True)
print("File 'dataset1.csv' has been updated as 'update_dataset1.csv'")
```
